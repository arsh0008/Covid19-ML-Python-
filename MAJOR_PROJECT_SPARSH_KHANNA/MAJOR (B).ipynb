```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv("covid_19_india .csv")
```

```python
data.drop('Sno',axis=1,inplace=True)
data.drop('ConfirmedForeignNational',axis=1,inplace=True)
data
```

```python
state = pd.DataFrame(data[data["State/UnionTerritory"]=="Punjab"])

state
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(state.corr(),annot=True,linewidth=0.5,linecolor="White")
```
